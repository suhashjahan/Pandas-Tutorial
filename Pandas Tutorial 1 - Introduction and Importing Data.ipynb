# What is Pandas?

**Pandas** is an open source tool for data analysis and manipulation built on top of the Python programming language. 

# What can be done with Pandas?

* Import datasets from spreadsheets, CSV files etc.
* Calculating summary statistics such as mean, median of columns or correlation between two columns.
* Flexible handling of missing data.
* Visualisation of data.


<br>Pandas built-in functions are able to do almost everything that you wish to do with your data.
